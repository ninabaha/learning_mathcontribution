蒙特卡罗法：把足够多的点撒在平面上，落在圆内的与总的落点的比例为圆面积和其外切正方形的比值。

蒙特卡罗法又称统计模拟法，使用随机数或·伪随机数来解决问题的方法，其所求的问题同一定的概率模型相联系。

原理就是当数据样本足够大的时候，事件发生的频率等于其概率。

蒙特卡罗主要是一种思想，其算法没有严格要求。可以用于求非线性规划的初始值。

```python
#求解圆周率
import numpy as np
import matplotlib.pyplot as plt
import random

p = 10000
r = 1
n = 0
x0, y0 = 1 ,1
plt.figure()
plt.xlabel('x')
plt.ylabel('y')
plt.title("monte kalo to pi")
for i in range(p):
    px, py = random.uniform(0,2), random.uniform(0,2)
    if (px-x0)**2 + (py-y0)**2 <= r**2:
        plt.plot(px,py,'.',color ='b')
        n +=1
    else:
        plt.plot(px,py,'.',color ='r')
plt.axis('equal')
plt.show()
s = (n/p)*4
print("pi=",s)
```

三门问题
你参加一档电视节目，节目组提供了A、B、C三扇门，主持人告诉你，其中一扇门后边有辆汽车，其他两扇门后面是一头山羊，你可以选择一扇门打开获得门后的东西。假如你选择了B们，这时，主持人打开了C门，让你看到C门后是只山羊，然后问你要不要改选A门?(你想要汽车)

改选问题，组织者排除了一个错误选项，此时的决策.

```python
#np.random.randint(low,high,size)  在low和high之间的整数组成矩阵
print(np.random.randint(1,6,(5,8)))
```

```python
import numpy as np
import random
p = 10000
a = 0#不换的选择
b = 0#换门
for i in range(p):
    x = np.random.randint(1,4)# 设定中奖的门
    y = np.random.randint(1,4)# 设定第一次选中的门
    if x == y:
        a +=1
    else:
        b +=1
print('根据蒙特卡罗算法不换选胜利的概率为：',a/p)
print('根据蒙特卡罗算法换选胜利的概率为：',b/p)
```
