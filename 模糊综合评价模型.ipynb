模糊集合的分类主要有：偏大型、偏小型、中间型，类似topsis的极大型、极小型、中间型。例如：年轻就是偏小型的模糊集合，年龄越小，隶属度越大；中年就是中间型。本质是隶属函数的单调性。

隶属函数的确定方法

1，模糊统计法；原理是找多个人对同一个模糊概念进行描述，用隶属频率去定义隶属度。时间有限，不采用

2.借助已有的客观尺度：借助已有的文献的指标去度量，例如将恩格尔系数用于小康家庭的定义。注意隶属度在0~1之间，可用归一化处理。

3.指派法：即凭主观意愿，在确定模糊集合所属的分类之后，指派一个隶属函数，比赛中常用，只需进行选择

还有德尔菲法，二元对比排序法，综合加权法等。

第四步是每一个评价都用上面所有因素如：[1，2，3，4]，然后多个评价人的评价组成模糊评价矩阵。

第五步，模糊矩阵的合成运算，通过合成算子，选择准确算法，转换成矩阵运算。即权重矩阵乘评价矩阵。

多层次模糊综合评价：就是每个因素还有下属的相关因素。例如分为二级评判，先评判第最下面的的因素，再评判上一级的因素，从外到内。

模糊综合评价的py只涉及到矩阵的乘法

```python
import numpy as np
# 影响运行费用的各因素的单因素评价矩阵为:
R23 = np.array([
[0.18,0.14,0.18,0.14,0.13,0.23],
[0.15,0.20,0.15,0.25,0.10,0.15],
[0.25,0.12,0.13,0.12,0.18,0.20],
[0.16,0.15,0.21,0.11,0.20,0.17],
[0.23,0.18,0.17,0.16,0.15,0.11],
[0.19,0.13,0.12,0.12,0.11,0.33],
[0.17,0.16,0.15,0.08,0.25,0.19]])
#权重分配为
A23 = np.array([0.20,0.15,0.10,0.10,0.20,0.15, 0.10])
# 评价结果
# np.dot是Numpy库中的一个函数，用于计算两个数组的点积。对于一维数组，它计算的是这两个数组的内积。
#对于二维数组(矩阵)，它计算的是矩阵乘法。
B23 = np.dot(A23,R23)
print(B23)
```
